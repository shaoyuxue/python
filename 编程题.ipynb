```python
import pandas as pd
import numpy as np
```

# 读取数据

```python
df=pd.read_excel(r'C:\Users\shaoyumei\Desktop\产品成本还原（面试）.xlsx')
df.head()
```

# 筛选出直接材料

```python
df_zc=df.query('成本项目名称=="直接材料"')
df_zc
```

# 第一题：求直接材料的合计值

```python
s1=df_zc.groupby('产品编码')['金额'].sum()
df1=pd.DataFrame({"直接材料合计":s1})
df1
```

# 第二题：按产品编码 分组计算直接材料的平均单价

```python
s2=df_zc.groupby('产品编码')['本期成本'].mean()
df2=pd.DataFrame({"直接材料平均单价":s2})
df2
```

# 筛选出间接材料

```python
df_jc=df.query('成本项目名称=="间接材料"').dropna()
df_jc
```

# 第二题：按产品编码 分组计算间接材料的平均单价

```python
s3=df_jc.groupby('产品编码')['本期成本'].mean()
df3=pd.DataFrame({"间接材料平均单价":s3})
df3
```

# 汇总结果并输出到Excel

```python
df_result=pd.concat([df1,df2,df3],axis=1)
df_result.loc['合计']=df_result.apply(lambda x:x.sum())
df_result=df_result.round(decimals=2)
df_result
```

```python
df_result.to_excel(r'C:\Users\shaoyumei\Desktop\输出结果.xlsx')
```
